```python
import requests, sys
import pandas as pd
 
server = "https://rest.ensembl.org"
```

```python
ext = "/info/analysis/homo_sapiens?"
r = requests.get(server+ext, headers=headers)
 
if not r.ok:
  r.raise_for_status()
  sys.exit()
 
decoded = r.json()
print(repr(decoded))
```

```python
#Endpoint: /overlap/region/:species/:region
# para obtener datos de genes, transcripciones y variaciones en una región del genoma humano:

ext = "/overlap/region/homo_sapiens/1:1000000-5000000?feature=gene"
 
r = requests.get(server+ext, headers={ "Content-Type" : "application/json"})
 
if not r.ok:
  r.raise_for_status()
  sys.exit()
 
decoded = r.json()
print(repr(decoded))

df_genoma_humano = pd.json_normalize(decoded)

df_genoma_humano.head(5)
```

```python
df_genoma_humano.info()
```

```python
#Endpoint: /overlap/region/:species/:region?feature=variation
# para obtener variantes genéticas en el cromosoma 1, entre las posiciones 1,000,000 y 5,000,000

ext = "/overlap/region/homo_sapiens/1:1000000-5000000?feature=gene"
 
r = requests.get(server+ext, headers={ "Content-Type" : "application/json"})
 
if not r.ok:
  r.raise_for_status()
  sys.exit()
 
decoded = r.json()
print(repr(decoded))

df_cromosoma = pd.json_normalize(decoded, sep='_')

df_cromosoma.head(5)
```

```python
df_cromosoma.info()
```

```python
#ENSG00000139618 es el ID de Ensembl para el gen BRCA2, 
# un gen humano importante que está asociado con la predisposición genética a varios tipos de cáncer, 
# incluyendo el cáncer de mama y ovario hereditario. 
# BRCA2 codifica una proteína que es crucial para la reparación del ADN.
ext = "/lookup/id/ENSG00000139618?expand=1"

r = requests.get(server + ext, headers=headers)

if not r.ok:
  r.raise_for_status()
  sys.exit()

decoded = r.json()

df_BRCA2 = pd.json_normalize(decoded)

df_BRCA2.head()
```

```python
ext = "/overlap/id/ENSG00000139618?feature=variation&limit=300&offset=0"

r = requests.get(server + ext, headers=headers)

if not r.ok:
  r.raise_for_status()
  sys.exit()

decoded = r.json()

df_BRCA2 = pd.json_normalize(decoded)

df_BRCA2.head()
```
